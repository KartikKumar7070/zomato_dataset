```python
ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
ratings.head()
```

obeservation
whenever the rating is from 4.5 to 4.9 this indicates that it is excellent
whenever the rating is from 4.0 to 4.5 this indicates that it is very good 
whenever the rating is from 3.5 to 3.9 this indicates that it is good
whenever the rating is from 3.0 to 3.4 this indicates that it is average
whenever the rating is from 2.5 to 2.9 this indicates that it is avergage
whenever the rating is from 2.4 to 0 this indicates that it is poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y= 'Rating Counts',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y= 'Rating Counts', hue='Rating color', data=ratings,pallete=['White','Red','Orange','Yellow','Green','Dark','Green'])
```

Observation
1. not rated count is very high
2. max number of rating are between 2.5 to 3.4

```python
sns.countplot(x= 'Rating color' data = ratings)
```
